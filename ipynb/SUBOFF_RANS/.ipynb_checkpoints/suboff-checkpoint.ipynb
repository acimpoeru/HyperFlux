```python
from zutil.post import cp_profile_wall_from_file

def plot_cp_profile(ax,file_root,symmetry,section_normal, section_loc):
    force_data = cp_profile_wall_from_file(file_root,
                                    
                                           section_normal,
                                           section_loc,
                                           func=plot_array,
                                           axis=ax,
                                           symmetry=symmetry,
                                           #section_data=section_data,
                                           alpha=0.0)
loop_counter=0
loop_data = {}
def plot_array(data_array,pts_array,**kwargs):
    global loop_counter, loop_data
    ax = kwargs['axis']
    symmetry = kwargs['symmetry']
    cp_array = data_array.GetPointData()['cp']
    chord_array = data_array.GetPointData()['chord']
    
    coordinates = pts_array.GetPoints()
    #symmetry.append((pts_array.GetPoints(),cp_array))
    loop_data[loop_counter] = (pts_array) 
    ax.plot(chord_array, cp_array , 'b.',color='k',label='zCFD SST',markersize=10)
    #print chord_array
```
